```python
!pip install --quiet easyocr opencv-python-headless numpy scikit-learn matplotlib tensorflow pandas


import os, sys, math, time, json, shutil, zipfile, string
from pathlib import Path
import numpy as np
import pandas as pd
import cv2
from PIL import Image
import matplotlib.pyplot as plt

import easyocr
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model

from google.colab import files
def show_image_with_ocr(path, ocr_results, title="VIN OCR Detection"):
    import cv2, matplotlib.pyplot as plt

    img = cv2.imread(str(path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    for (bbox, text, conf) in ocr_results:
        pts = np.array(bbox, dtype=int)
        cv2.polylines(img_rgb, [pts], isClosed=True, color=(0,255,0), thickness=2)
        for (x,y) in pts:
            cv2.circle(img_rgb, (x,y), 5, (255,0,0), -1)

    plt.figure(figsize=(10,4))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(title)
    plt.show()


IMG_SIZE = (64, 64)
ALLOWLIST = string.ascii_uppercase + string.digits
TEMPLATES_DIR = Path("/content/templates")
DATASET_ROOT = Path("/content/dataset")
CNN_MODEL_PATH = Path("/content/tresnet_font_model.h5")
CNN_LABELS_PATH = Path("/content/label_names.txt")

OUTPUT_CSV = "vin_analysis_results.csv"

MIN_VIN_LEN = 10
TAMPER_DECISION_THRESHOLD = 50.0
VIN_CHARS = set(ALLOWLIST.replace("O", ""))


def ensure_dir(p: Path):
    p.parent.mkdir(parents=True, exist_ok=True)

def read_image_gray(path):
    img = cv2.imread(str(path), cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Cannot read image: {path}")
    return img

def to_rgb(img_bgr):
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

def save_debug_image(path, img):
    ensure_dir(Path(path))
    cv2.imwrite(str(path), img)
def show_image_inline(img_path, title="VIN Image"):
    import cv2, matplotlib.pyplot as plt
    img = cv2.imread(str(img_path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,4))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(title)
    plt.show()


def levenshtein(a, b):
    if a == b: return 0
    if len(a) == 0: return len(b)
    if len(b) == 0: return len(a)
    dp = np.zeros((len(a)+1, len(b)+1), dtype=int)
    for i in range(len(a)+1): dp[i,0] = i
    for j in range(len(b)+1): dp[0,j] = j
    for i in range(1, len(a)+1):
        for j in range(1, len(b)+1):
            cost = 0 if a[i-1] == b[j-1] else 1
            dp[i,j] = min(dp[i-1,j] + 1, dp[i,j-1] + 1, dp[i-1,j-1] + cost)
    return dp[len(a), len(b)]

def char_accuracy(pred, gt):
    if not gt:
        return None
    if len(gt) == 0:
        return None
    dist = levenshtein(pred, gt)
    return max(0.0, 1.0 - dist / max(len(gt), 1)) * 100.0

def normalize_conf(c):
    if c is None: return 0.0
    return float(np.clip(c, 0.0, 1.0) * 100.0)

def safe_ratio(a, b):
    return 0.0 if b == 0 else float(a) / float(b)


WMI_MAP = {
    "1HG": "Honda USA", "JHM": "Honda Japan", "WVW": "Volkswagen Germany",
    "WDB": "Mercedes-Benz Germany", "WBA": "BMW Germany", "WDC": "Mercedes-Benz Germany",
    "3VW": "Volkswagen Mexico", "SAL": "Land Rover UK", "MA1": "Mahindra India",
    "MAL": "Hyundai Korea", "KMH": "Hyundai Korea", "KNM": "Kia Korea",
    "MH1": "Honda India", "MRH": "Honda Thailand", "KL1": "GM Korea",
}

def wmi_brand_from_vin(vin: str):
    vin = "".join(ch for ch in vin.upper() if ch in ALLOWLIST)
    if len(vin) < 3:
        return "Unknown"
    wmi = vin[:3]
    return WMI_MAP.get(wmi, "Unknown")


class FontCNN:
    def __init__(self, model_path: Path, labels_path: Path):
        self.ok = False
        self.model = None
        self.labels = []
        if model_path.exists() and labels_path.exists():
            try:
                self.model = load_model(str(model_path))
                with open(labels_path, "r") as f:
                    self.labels = [ln.strip() for ln in f if ln.strip()]
                self.ok = True
            except Exception as e:
                print(f"[WARN] Could not load CNN model: {e}")

    def predict(self, img_path: Path):
        if not self.ok:
            return ("N/A", 0.0, None)
        pil = Image.open(str(img_path)).convert('L').resize(IMG_SIZE)
        arr = img_to_array(pil) / 255.0
        arr = np.expand_dims(arr, axis=0)
        probs = self.model.predict(arr, verbose=0)[0]
        idx = int(np.argmax(probs))
        label = self.labels[idx] if idx < len(self.labels) else f"class_{idx}"
        conf = float(probs[idx])
        return (label, conf, probs)

font_cnn = FontCNN(CNN_MODEL_PATH, CNN_LABELS_PATH)


_eocr_reader = None
def get_reader():
    global _eocr_reader
    if _eocr_reader is None:
        _eocr_reader = easyocr.Reader(['en'], gpu=False, verbose=False)
    return _eocr_reader

def easyocr_detect_vin(img_path: Path):
    reader = get_reader()
    img = cv2.imread(str(img_path))
    if img is None:
        return "", 0.0

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.bilateralFilter(gray, 5, 40, 40)
    gray = cv2.equalizeHist(gray)

    results = reader.readtext(gray, detail=1, allowlist=ALLOWLIST, paragraph=False, contrast_ths=0.05, text_threshold=0.4)

    if not results:
        return "", 0.0

    results = sorted(results, key=lambda x: min(pt[0] for pt in x[0]))

    vin_fragments = []
    confs = []
    for bbox, text, conf in results:
        t = "".join(ch for ch in text.upper() if ch in ALLOWLIST)
        if len(t) >= 2:
            vin_fragments.append(t)
            confs.append(conf)

    vin_text = "".join(vin_fragments)
    vin_conf = np.mean(confs) if confs else 0.0

    vin_text = vin_text.replace("O", "0")

    return vin_text, float(vin_conf)



def swt_font_detection(img_path: Path):
    img = read_image_gray(str(img_path))

    v = np.median(img)
    low = int(max(0, 0.66 * v))
    high = int(min(255, 1.33 * v))
    edges = cv2.Canny(img, low, high)

    dist = cv2.distanceTransform(255 - edges, cv2.DIST_L2, 5)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    mask = cv2.morphologyEx(255 - edges, cv2.MORPH_OPEN, kernel)
    pos = dist[(dist > 0) & (mask > 0)]
    if pos.size == 0:
        return ("Unknown", 0.0, 0.0)

    med_sw = float(np.median(pos))
    iqr = float(np.percentile(pos, 75) - np.percentile(pos, 25) + 1e-6)
    consistency = float(np.clip(1.0 / (1.0 + (iqr / (med_sw + 1e-6))), 0.0, 1.0))


    if med_sw < 2.0:
        label = "dot peened"
    elif med_sw < 4.0:
        label = "ocrb" if consistency > 0.8 else "engraved"
    elif med_sw < 7.0:
        label = "engraved"
    else:
        label = "embossed"

    d = 0.0
    if label == "dot peened":
        d = max(0.0, (2.0 - med_sw) / 2.0)
    elif label in ("ocrb", "engraved"):
        center = 3.0
        d = max(0.0, 1.0 - abs(med_sw - center) / 1.5)
    else:
        d = max(0.0, (med_sw - 7.0) / 7.0)
    conf = float(np.clip(0.5 * consistency + 0.5 * d, 0.0, 1.0))
    return (label, conf, med_sw)


def contour_matching(img_path: Path, templates_dir: Path):
    img = read_image_gray(str(img_path))
    img = cv2.GaussianBlur(img, (3,3), 0)
    img_edges = cv2.Canny(img, 60, 160)

    candidates = []
    scores = {}
    classes = ["dot peened", "engraved", "embossed", "ocrb"]
    for cls in classes:
        tpl_path = templates_dir / f"{cls.replace(' ', '_')}_template.jpg"
        if not tpl_path.exists():
            continue
        tpl = read_image_gray(str(tpl_path))
        tpl = cv2.resize(tpl, (img.shape[1], img.shape[0]))
        tpl = cv2.GaussianBlur(tpl, (3,3), 0)
        tpl_edges = cv2.Canny(tpl, 60, 160)

        res = cv2.matchTemplate(img_edges, tpl_edges, cv2.TM_CCOEFF_NORMED)
        score = float(res.max())
        scores[cls] = score
        candidates.append((cls, score))

    if not candidates:
        return ("N/A", 0.0, {})

    candidates.sort(key=lambda x: x[1], reverse=True)
    best_cls, best_score = candidates[0]
    conf = np.clip((best_score + 1.0) / 2.0, 0.0, 1.0)
    return (best_cls, float(conf), scores)


def edge_and_corners_v2(img_path: Path, save_base: str):
    img = cv2.imread(str(img_path))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 60, 160)

    corners = cv2.goodFeaturesToTrack(gray, maxCorners=200, qualityLevel=0.01, minDistance=5)
    if corners is None:
        corners = []
    else:
        corners = corners.astype(int)

    out = np.zeros_like(img)
    out[edges > 0] = [255, 255, 255]
    for c in corners:
        x, y = c.ravel()
        cv2.circle(out, (x, y), 3, (0, 0, 255), -1)

    edges_path = f"{save_base}_edges.jpg"
    viz_path = f"{save_base}_analysis.jpg"
    save_debug_image(edges_path, (edges > 0).astype(np.uint8) * 255)
    save_debug_image(viz_path, out)
    return edges_path, viz_path


def final_decision(
    ocr_score_pct,
    cnn_label, cnn_conf,
    swt_label, swt_conf,
    gt_font=None
):
    font_agree = 0.0
    if gt_font:
        font_agree = 100.0 if (cnn_label.lower() == gt_font.lower() or swt_label.lower() == gt_font.lower()) else 0.0
    else:
        if cnn_label != "N/A" and swt_label != "Unknown":
            font_agree = 100.0 if cnn_label.lower() == swt_label.lower() else 0.0
        else:
            font_agree = 50.0

    cnn_score = normalize_conf(cnn_conf)
    swt_score = normalize_conf(swt_conf)

    disagree = 0
    if cnn_label != "N/A" and swt_label != "Unknown" and cnn_label and swt_label:
        disagree = 1 if cnn_label.lower() != swt_label.lower() else 0
    font_irregularity = float(np.clip((disagree * 60.0 + (100.0 - max(cnn_score, swt_score)) * 0.4), 0.0, 100.0))

    tamper_raw = (
        (100.0 - float(ocr_score_pct)) * 0.55 +
        (100.0 - font_agree) * 0.25 +
        (font_irregularity) * 0.20
    )
    tamper_raw = float(np.clip(tamper_raw, 0.0, 100.0))
    genuine_pct = float(np.clip(100.0 - tamper_raw, 0.0, 100.0))
    verdict = "Genuine" if genuine_pct >= TAMPER_DECISION_THRESHOLD else "Tempered"

    return {
        "cnn_score_pct": cnn_score,
        "swt_score_pct": swt_score,
        "font_agree_pct": font_agree,
        "font_irregularity_pct": font_irregularity,
        "tamper_raw_pct": tamper_raw,
        "genuine_pct": genuine_pct,
        "verdict": verdict
    }


all_records = []

print(" Upload one or more images. To finish, cancel the chooser.")
while True:
    try:
        uploaded = files.upload()
        if not uploaded:
            break
    except Exception:
        break

    for fname in uploaded.keys():
        path = Path(fname)
        print(f"\nProcessing: {fname}")

        reader = get_reader()  # initialize the EasyOCR reader
        ocr_results = reader.readtext(str(path), allowlist=ALLOWLIST)

        ocr_text = "".join([t for (_,t,_) in ocr_results]).upper()

        show_image_with_ocr(path, ocr_results)

        gt_vin = input("1) Enter Ground Truth VIN (or press Enter to skip): ").strip().upper()
        gt_font = input("2) Enter Ground Truth Font (engraved/embossed/dot peened/ocrb) (or press Enter): ").strip().lower()

        t0 = time.time()
        ocr_text, ocr_conf = easyocr_detect_vin(path)
        ocr_text = ocr_text.upper()
        ocr_char_acc = char_accuracy(ocr_text, gt_vin) if gt_vin else None
        ocr_score_pct = ocr_char_acc if ocr_char_acc is not None else normalize_conf(ocr_conf)

        cnn_label, cnn_conf, _ = font_cnn.predict(path)
        if cnn_label != "N/A":
            cnn_label_norm = cnn_label.lower()
        else:
            cnn_label_norm = "N/A"

        swt_label, swt_conf, sw_med = swt_font_detection(path)
        swt_label_norm = swt_label.lower()

        contour_label = "N/A"
        contour_score_pct = 0.0
        contour_scores = {}
        if TEMPLATES_DIR.exists():
            contour_label, contour_conf, contour_scores = contour_matching(path, TEMPLATES_DIR)
            contour_score_pct = normalize_conf(contour_conf)

        base = path.stem
        edges_path, viz_path = edge_and_corners_v2(path, save_base=base)

        brand = wmi_brand_from_vin(gt_vin if gt_vin else ocr_text)

        decision = final_decision(
            ocr_score_pct=ocr_score_pct if ocr_score_pct is not None else 0.0,
            cnn_label=cnn_label_norm, cnn_conf=cnn_conf,
            swt_label=swt_label_norm, swt_conf=swt_conf,
            gt_font=gt_font if gt_font else None
        )

        t1 = time.time()

        print("\n=== OUTPUT (1..11) ===")
        print(f"1: GROUND TRUTH INPUT OF VIN: {gt_vin if gt_vin else 'N/A'}")
        conf_str = f"{ocr_conf:.3f}"
        print(f"2: EASYOCR VIN DETECTION: {ocr_text if ocr_text else 'N/A'} (EasyOCR conf {conf_str})")
        print(f"3: GROUND TRUTH OF FONT: {gt_font if gt_font else 'N/A'}")
        print("4: FONT DETECTION:")
        print(f"   - CNN predicted font: {cnn_label if cnn_label != 'N/A' else 'N/A'} ({normalize_conf(cnn_conf):.2f}%)")
        print(f"   - SWT predicted font: {swt_label} ({normalize_conf(swt_conf):.2f}% )")
        print("5: ACCURACY:")
        if ocr_char_acc is not None:
            print(f"   - OCR accuracy (%): {ocr_char_acc:.2f}")
        else:
            print("   - OCR accuracy (%): N/A (no VIN ground truth)")
        if gt_font:
            print(f"   - CNN font match GT: {str(cnn_label_norm == gt_font.lower()) if cnn_label != 'N/A' else 'False'}")
            print(f"   - SWT font match GT: {str(swt_label_norm == gt_font.lower())}")
        else:
            print("   - CNN font match GT: N/A (no font ground truth)")
            print("   - SWT font match GT: N/A (no font ground truth)")
        print(f"6: SWT ACCURACY (score 0..100): {normalize_conf(swt_conf):.2f}")
        if contour_label != "N/A":
            print("7: CONTOUR MATCHING (per-class scores):")
            for k, v in contour_scores.items():
                print(f"   - {k}: {normalize_conf((v + 1.0)/2.0):.2f}%")
        else:
            print("7: CONTOUR MATCHING (per-class scores):")
            print("   - No templates available (no dataset) or no matches.")
        print(f"8: EDGE DETECTION (image saved) with VIN corners highlighted red -> {viz_path} and edges image: {edges_path}")
        print(f"9: WMI DETECTION (brand): {brand}")
        print("10: Calculations & extracted results and combined average:")
        print(f"    - CNN score (0..100): {decision['cnn_score_pct']:.2f}")
        print(f"    - SWT score (0..100): {decision['swt_score_pct']:.2f}")
        print(f"    - OCR score (0..100): {ocr_score_pct:.2f}")
        print(f"    - Tamper raw score (0..100, higher=more tampered): {decision['tamper_raw_pct']:.2f}")
        print(f"    - Tamper inverted (genuine-like): {100.0 - decision['tamper_raw_pct']:.2f}")
        print(f"    - Font irregularity (heuristic 0..100): {decision['font_irregularity_pct']:.2f}")
        print(f"    - Combined final confidence (genuine %): {decision['genuine_pct']:.2f}")
        print(f"11: PREDICTION OF VIN IS TEMPERED OR GENUINE -> {decision['verdict']}")
        print(f"    - Processing time (s): {t1 - t0:.2f}")

        rec = {
            "filename": fname,
            "gt_vin": gt_vin,
            "ocr_vin": ocr_text,
            "ocr_conf_est": ocr_conf,
            "ocr_score_pct": ocr_score_pct,
            "gt_font": gt_font,
            "cnn_font": cnn_label,
            "cnn_conf_pct": decision['cnn_score_pct'],
            "swt_font": swt_label,
            "swt_conf_pct": decision['swt_score_pct'],
            "contour_font": contour_label,
            "contour_conf_pct": contour_score_pct,
            "brand_wmi": brand,
            "tamper_raw_pct": decision['tamper_raw_pct'],
            "genuine_pct": decision['genuine_pct'],
            "verdict": decision['verdict'],
            "edges_img": edges_path,
            "analysis_img": viz_path
        }
        all_records.append(rec)


if all_records:
    df = pd.DataFrame(all_records)
    df.to_csv(OUTPUT_CSV, index=False)
    print(f"\n‚úÖ Results saved to {OUTPUT_CSV}")
else:
    print("\nNo images processed. Nothing to save.")
```

```python
import json
import re
from dataclasses import dataclass
from typing import List, Dict, Any, Optional
from enum import Enum
from datetime import datetime

# ---------------------------
# Risk levels
# ---------------------------
class RiskLevel(Enum):
    PASS = "PASS"
    REVIEW = "REVIEW"
    SUSPECT = "SUSPECT"
    HIGH_RISK = "HIGH_RISK"

@dataclass
class CEOResult:
    vin: str
    risk_level: RiskLevel
    score: int
    confidence: float
    suggested_action: str
    reasons: List[str]
    details: Dict[str, Any]
    timestamp: str

class CEOFraudEngine:
    def __init__(self):
        # Small WMI database (extendable)
        self.wmi_db = {
            "1HG": "Honda USA", "JHM": "Honda Japan", "WVW": "Volkswagen Germany",
            "WDB": "Mercedes-Benz Germany", "WBA": "BMW Germany", "WDC": "Mercedes-Benz Germany",
            "3VW": "Volkswagen Mexico", "SAL": "Land Rover UK", "MA1": "Mahindra India",
            "MAL": "Hyundai Korea", "KMH": "Hyundai Korea", "KNM": "Kia Korea",
            "MH1": "Honda India", "MRH": "Honda Thailand", "KL1": "GM Korea",
        }

    def clean_vin(self, vin: str) -> str:
        if not vin:
            return ""
        return vin.strip().upper()

    def level1(self, vin: str, doc_vin: str, wmi_brand: Optional[str], details: Dict[str, Any]):
        score = 0
        reasons = []

        vin_clean = self.clean_vin(vin)
        if len(vin_clean) == 17:
            score += 8
            reasons.append("VIN length 17 (format OK).")
        elif 11 <= len(vin_clean) <= 16:
            score += 2
            reasons.append(f"Non-standard VIN length {len(vin_clean)} (JDM/short).")
        else:
            score -= 12
            reasons.append(f"Invalid VIN length {len(vin_clean)} (format fail).")

        # Forbidden chars
        if any(c in vin_clean for c in ["I", "O", "Q"]):
            score -= 12
            reasons.append("VIN contains forbidden letters (I/O/Q).")

        wmi = vin_clean[:3] if len(vin_clean) >= 3 else ""
        if wmi_brand and wmi_brand.lower() != "unknown":
            score += 6
            reasons.append(f"WMI detection: {wmi_brand}")
        elif wmi in self.wmi_db:
            score += 6
            reasons.append(f"WMI lookup OK: {self.wmi_db[wmi]}")
        else:
            score -= 6
            reasons.append(f"WMI unknown: {wmi}")

        if doc_vin and vin_clean and doc_vin.strip().upper() == vin_clean:
            score += 8
            reasons.append("VIN matches paperwork/documents.")
        else:
            if doc_vin:
                score -= 15
                reasons.append(f"VIN mismatch with documents: doc={doc_vin.strip().upper()} vs VIN={vin_clean}")
            else:
                score -= 8
                reasons.append("No VIN found on paperwork (documents missing).")

        year_note = ""
        if len(vin_clean) >= 10:
            year_char = vin_clean[9]
            year_map = {
                'A': 2010, 'B': 2011, 'C': 2012, 'D': 2013, 'E': 2014,
                'F': 2015, 'G': 2016, 'H': 2017, 'J': 2018, 'K': 2019,
                'L': 2020, 'M': 1991, 'N': 1992, 'P': 1993, 'R': 1994,
                'S': 1995, 'T': 1996, 'V': 1997, 'W': 1998, 'X': 1999,
                'Y': 2000, '1': 2001, '2': 2002, '3': 2003, '4': 2004,
                '5': 2005, '6': 2006, '7': 2007, '8': 2008, '9': 2009
            }
            if year_char in year_map:
                year_note = f"VIN year code {year_char} -> approx {year_map[year_char]}"
                score += 2
                reasons.append(year_note)
            else:
                reasons.append(f"VIN year code {year_char} unknown.")
        else:
            reasons.append("VIN too short to derive year code.")

        details.update({"level1_wmi": wmi, "level1_year_note": year_note})
        return score, reasons

    def level2(self, vin: str, chassis_stamp: Optional[str], hidden_vins: List[str], engine_number: Optional[str]):
        score = 0
        reasons = []

        vin_c = self.clean_vin(vin)
        locations = [x for x in [vin_c, self.clean_vin(chassis_stamp)] if x]
        unique = set(locations)

        if len(unique) == 1 and len(locations) >= 2:
            score += 10
            reasons.append("Visible VINs (plate/chassis) consistent across locations.")
        elif len(locations) >= 2:
            score -= 20
            reasons.append(f"Mismatch across visible VIN locations: {unique}")
        else:
            score += 2
            reasons.append("Single visible VIN location (no cross-plate check).")

        if hidden_vins:
            matches = sum(1 for h in hidden_vins if self.clean_vin(h) == vin_c)
            if matches == len(hidden_vins):
                score += 12
                reasons.append(f"All {len(hidden_vins)} hidden secondary VINs match the visible VIN.")
            elif matches > 0:
                score -= 10
                reasons.append(f"Some hidden VINs differ from visible VIN ({matches}/{len(hidden_vins)} match).")
            else:
                score -= 25
                reasons.append("Hidden VINs present but none match visible VIN.")
        else:
            score -= 8
            reasons.append("No hidden VINs checked (not provided).")

        if engine_number and len(engine_number.strip()) > 0:
            score += 6
            reasons.append("Engine number present (documented).")
        else:
            score -= 8
            reasons.append("Engine number missing or not documented.")

        return score, reasons

    def level3(self, temper: Dict[str, Any]):
        score = 0
        reasons = []
        details = {}

        ocr = float(temper.get("ocr_score", 0.0))
        details["ocr_score"] = ocr
        if ocr >= 90:
            score += 10
            reasons.append(f"OCR accuracy excellent ({ocr:.2f}%).")
        elif ocr >= 75:
            score += 6
            reasons.append(f"OCR accuracy good ({ocr:.2f}%).")
        elif ocr >= 50:
            score += 2
            reasons.append(f"OCR accuracy fair ({ocr:.2f}%).")
        else:
            score -= 8
            reasons.append(f"OCR accuracy poor ({ocr:.2f}%).")

        swt = float(temper.get("swt_score", 0.0))
        cnn = float(temper.get("cnn_score", 0.0))
        details["swt_score"] = swt
        details["cnn_score"] = cnn

        font_gt = temper.get("ground_truth_font", None)
        swt_label = temper.get("swt_label", None)
        cnn_label = temper.get("cnn_label", None)

        if swt >= 50 and (font_gt is None or (swt_label and font_gt.lower() in swt_label.lower())):
            score += 6
            reasons.append(f"SWT font detection confident ({swt:.2f}), label: {swt_label}.")
        else:
            if swt_label:
                reasons.append(f"SWT font weak ({swt:.2f}) label: {swt_label}.")
            else:
                reasons.append("SWT font not decisive.")
            score -= 2

        if cnn >= 80 and swt >= 50 and cnn_label and swt_label and cnn_label.lower() == swt_label.lower():
            score += 6
            reasons.append(f"CNN ({cnn:.2f}) & SWT agree on font ({cnn_label}).")
        elif cnn >= 80 and (not swt or swt < 50):
            score += 3
            reasons.append(f"CNN confident font ({cnn_label}) but SWT weak.")
        else:
            reasons.append("No strong CNN font confirmation.")

        tamper_raw = float(temper.get("tamper_raw", 0.0))
        details["tamper_raw"] = tamper_raw
        if tamper_raw < 30:
            score += 8
            reasons.append(f"Tamper raw low ({tamper_raw:.2f}) ‚Üí likely genuine surface.")
        elif tamper_raw < 50:
            score += 0
            reasons.append(f"Tamper raw borderline ({tamper_raw:.2f}) ‚Üí manual review recommended.")
        else:
            score -= 18
            reasons.append(f"Tamper raw high ({tamper_raw:.2f}) ‚Üí strong tampering indicator.")

        font_irreg = float(temper.get("font_irregularity", 0.0))
        details["font_irregularity"] = font_irreg
        if font_irreg >= 70:
            score -= 8
            reasons.append(f"High font irregularity ({font_irreg:.2f}) ‚Üí suspicious.")
        elif font_irreg >= 50:
            score -= 2
            reasons.append(f"Moderate font irregularity ({font_irreg:.2f}).")
        else:
            score += 2
            reasons.append(f"Low font irregularity ({font_irreg:.2f}).")

        rivets = temper.get("rivet_count", None)
        if rivets is not None:
            details["rivet_count"] = rivets
            try:
                r = int(rivets)
                if 2 <= r <= 4:
                    score += 4
                    reasons.append(f"Rivet count normal ({r}).")
                else:
                    score -= 6
                    reasons.append(f"Rivet count abnormal ({r}).")
            except Exception:
                reasons.append("Rivet count provided but invalid.")

        return score, reasons, details

    def level4(self, temper: Dict[str, Any], vin_year_hint: Optional[int] = None, part_dates: Optional[Dict[str,int]] = None):
        score = 0
        reasons = []
        if part_dates:
            inconsistencies = 0
            for comp, year in part_dates.items():
                if vin_year_hint:
                    diff = abs(year - vin_year_hint)
                    if diff <= 1:
                        score += 2
                        reasons.append(f"{comp} date {year} matches VIN year {vin_year_hint}.")
                    elif diff <= 2:
                        reasons.append(f"{comp} date {year} slightly off VIN year {vin_year_hint}.")
                    else:
                        score -= 8
                        inconsistencies += 1
                        reasons.append(f"{comp} date {year} inconsistent with VIN year {vin_year_hint}.")
                else:
                    reasons.append(f"{comp} date {year} noted (no VIN-year available).")
            if inconsistencies == 0 and len(part_dates) >=2:
                score += 8
                reasons.append("Component date codes coherent.")
            elif inconsistencies >= 2:
                score -= 15
                reasons.append("Multiple component date inconsistencies.")
        else:
            reasons.append("No component date codes provided (level 4 skipped).")

        vin = temper.get("ground_truth_vin", "")
        if len(vin) >= 11:
            plant_char = vin[10]
            if plant_char.isalnum():
                score += 2
                reasons.append(f"Plant code format OK ('{plant_char}').")
            else:
                score -= 4
                reasons.append(f"Plant code invalid ('{plant_char}').")
        else:
            reasons.append("VIN too short for plant code check.")

        return score, reasons

    def level5(self, temper: Dict[str, Any]):
        score = 0
        reasons = []
        auction = temper.get("auction_photos_match", None)
        theft_clear = temper.get("theft_check_clear", None)

        if auction is True:
            score += 10
            reasons.append("Auction/import photos match (provenance positive).")
        elif auction is False:
            score -= 20
            reasons.append("Auction/import photos do not match (provenance negative).")
        else:
            reasons.append("No auction provenance provided.")

        if theft_clear is True:
            score += 6
            reasons.append("Theft database check clear.")
        elif theft_clear is False:
            score -= 25
            reasons.append("Vehicle flagged in theft database.")
        else:
            reasons.append("No theft database check provided.")

        return score, reasons

    # Decision thresholds
    def decide(self, total_score: int, combined_confidence: float):
        if total_score >= 25:
            return RiskLevel.PASS, max(0.5, min(0.98, combined_confidence)), "Vehicle passes fraud screening (low risk)."
        elif 10 <= total_score <= 24:
            return RiskLevel.REVIEW, max(0.4, min(0.9, combined_confidence*0.9)), "Recommend manual review / additional verification."
        elif -5 <= total_score <= 9:
            return RiskLevel.SUSPECT, max(0.35, min(0.85, combined_confidence*0.7)), "Suspicious - recommend forensic inspection / in-person check."
        else:
            return RiskLevel.HIGH_RISK, max(0.6, min(0.99, (1 - combined_confidence) if combined_confidence <= 1 else 0.9)), "High probability of tampering - reject or escalate to forensic."

    # Main evaluate method
    def evaluate_sample(self, temper: Dict[str, Any], chassis_stamp: Optional[str] = None, hidden_vins: Optional[List[str]] = None, engine_number: Optional[str] = None, doc_vin: Optional[str] = None, part_dates: Optional[Dict[str,int]] = None) -> CEOResult:
        vin = temper.get("ground_truth_vin") or temper.get("vin") or ""
        vin = self.clean_vin(vin)
        reasons_total = []
        details_total = {"temper": temper.copy()}

        # Level 1
        l1_score, l1_reasons = self.level1(vin, doc_vin or vin, temper.get("wmi_brand"), details_total)
        reasons_total += [f"L1: {r}" for r in l1_reasons]

        # Level 2
        hidden_vins = hidden_vins or temper.get("hidden_vins", []) or []
        l2_score, l2_reasons = self.level2(vin, chassis_stamp or vin, hidden_vins, engine_number or temper.get("engine_number"))
        reasons_total += [f"L2: {r}" for r in l2_reasons]

        # Level 3 (ML / visual)
        l3_score, l3_reasons, l3_details = self.level3(temper)
        details_total.update(l3_details)
        reasons_total += [f"L3: {r}" for r in l3_reasons]

        # Level 4
        l4_score, l4_reasons = self.level4(temper, vin_year_hint=None, part_dates=part_dates)
        reasons_total += [f"L4: {r}" for r in l4_reasons]

        # Level 5
        l5_score, l5_reasons = self.level5(temper)
        reasons_total += [f"L5: {r}" for r in l5_reasons]

        total_score = int(round(l1_score + l2_score + l3_score + l4_score + l5_score))
        combined_conf = float(temper.get("combined_confidence", temper.get("combined_final_conf", temper.get("combined_conf", temper.get("combined_final_confidence", 0.75)))))

        risk_level, confidence, action = self.decide(total_score, combined_conf)
        # Assemble final result
        result = CEOResult(
            vin=vin,
            risk_level=risk_level,
            score=total_score,
            confidence=round(confidence, 4),
            suggested_action=action,
            reasons=reasons_total,
            details=details_total,
            timestamp=datetime.now().isoformat()
        )
        return result

# ---------------------------
# Example batch: Temper outputs
# ---------------------------
TEMPER_SAMPLES = [
  {
    "ground_truth_vin": "SB308PK012558787",
    "easyaocr_detect": "SB308PK0125878",
    "easyaocr_conf": 0.375,
    "ground_truth_font": "engraved",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 67.41,
    "ocr_score": 87.50,
    "tamper_raw": 34.48,
    "font_irregularity": 13.03,
    "combined_confidence": 0.6552,
    "wmi_brand": "Unknown",
    "prediction": "Genuine"
  },
  {
    "ground_truth_vin": "SB308PK012558787",
    "easyaocr_detect": "SUZUXIJMT0RC0RP0RATI0NTYPEENTGINECHASSISN0ENGINEN0C0L0R22S7RFKI24F8BPK72029401258787SB308796CCJAPAN",
    "easyaocr_conf": 0.728,
    "ground_truth_font": "dot peened",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 100.00,
    "ocr_score": 0.00,
    "tamper_raw": 80.00,
    "font_irregularity": 0.00,
    "combined_confidence": 0.20,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  },
  {
    "ground_truth_vin": "NFBFC6661JR046248",
    "easyaocr_detect": "NFBFC6661JR0Z6248CHSSISNHAAISCSIPKIS0LI47",
    "easyaocr_conf": 0.428,
    "ground_truth_font": "dot peened",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 100.00,
    "ocr_score": 0.00,
    "tamper_raw": 80.00,
    "font_irregularity": 0.00,
    "combined_confidence": 0.20,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  },
  {
    "ground_truth_vin": "NSP150R7057928",
    "easyaocr_detect": "01",
    "easyaocr_conf": 0.103,
    "ground_truth_font": "engraved",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 22.41,
    "ocr_score": 7.14,
    "tamper_raw": 82.28,
    "font_irregularity": 31.04,
    "combined_confidence": 0.1772,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  },
  {
    "ground_truth_vin": "NSP150R7057928",
    "easyaocr_detect": "M0DELMBPIS0REWPENGINEFRAMEN0PLANTTRANST0Y0TAM0T0RC0RP0RATI0NMFDBY3INDUSM0T0RC0MPANYLIMITED0K070BUILTAXLENSIS0R255C0L0R218K4J1020MRFETRIM9057928GVMKGIMADEINPAKUSTAN",
    "easyaocr_conf": 0.694,
    "ground_truth_font": "engraved",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 19.05,
    "ocr_score": 0.00,
    "tamper_raw": 86.48,
    "font_irregularity": 32.38,
    "combined_confidence": 0.1352,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  },
  {
    "ground_truth_vin": "NFBFC6661JR046248",
    "easyaocr_detect": "1L",
    "easyaocr_conf": 0.093,
    "ground_truth_font": "engraved",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "engraved",
    "swt_score": 42.19,
    "ocr_score": 5.88,
    "tamper_raw": 56.39,
    "font_irregularity": 23.12,
    "combined_confidence": 0.4361,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  },
  {
    "ground_truth_vin": "NSP150R7828697",
    "easyaocr_detect": "PLANTFRAMEN0TRANSENGINELNRFFEM0DELMFDBYEINDUSM0T0RC0MPANYLIMITED50K070I0DEY0TAM0T0RC0RP0RATI0NPAXLEBUILT255C542010CS2L50R67028697C0L0RNSRIE0RFIEMDKPTRIMFB00GVMKGD1329MADEINPAKISTANML",
    "easyaocr_conf": 0.745,
    "ground_truth_font": "engraved",
    "cnn_label": "N/A",
    "cnn_score": 0.00,
    "swt_label": "embossed",
    "swt_score": 100.00,
    "ocr_score": 0.00,
    "tamper_raw": 80.00,
    "font_irregularity": 0.00,
    "combined_confidence": 0.20,
    "wmi_brand": "Unknown",
    "prediction": "Tempered"
  }
]


# ---------------------------
# Run batch evaluation
# ---------------------------
# ---------------------------
# Run batch evaluation with engaging output
# ---------------------------
def run_batch(samples: List[Dict[str, Any]]):
    engine = CEOFraudEngine()
    reports: List[Dict[str, Any]] = []

    # Risk level icons/colors
    level_icons = {
        "PASS": "‚úÖ PASS (Green)",
        "REVIEW": "‚ö†Ô∏è REVIEW (Yellow)",
        "SUSPECT": "‚ùå SUSPECT (Orange)",
        "HIGH_RISK": "üö® HIGH RISK (Red)"
    }

    def confidence_bar(conf: float, length: int = 20) -> str:
        filled = int(conf * length)
        return "‚ñì" * filled + "‚ñë" * (length - filled)

    for i, s in enumerate(samples, start=1):
        report_obj = engine.evaluate_sample(
            temper=s,
            chassis_stamp=s.get("easyaocr_detect"),
            hidden_vins=s.get("hidden_vins", []),
            engine_number=s.get("engine_number", ""),
            doc_vin=s.get("ground_truth_vin"),
            part_dates=None
        )

        print("\n" + "="*70)
        print(f"üìÑ SAMPLE #{i}")
        print(f"üîë VIN           : {report_obj.vin}")
        print(f"üìä Risk Level    : {level_icons[report_obj.risk_level.value]}")
        print(f"üìà Fraud Score   : {report_obj.score}")
        print(f"üîí Confidence    : {report_obj.confidence:.0%}  [{confidence_bar(report_obj.confidence)}]")
        print(f"üìù Suggested Act.: {report_obj.suggested_action}")
        print("-"*70)

        # Group reasons by level
        grouped = {"L1": [], "L2": [], "L3": [], "L4": [], "L5": []}
        for r in report_obj.reasons:
            for lvl in grouped:
                if r.startswith(lvl):
                    grouped[lvl].append(r.replace(f"{lvl}: ", ""))

        for lvl, reasons in grouped.items():
            if reasons:
                print(f"üîç {lvl} Findings:")
                for rr in reasons:
                    print(f"   ‚Ä¢ {rr}")
                print("-"*70)

        print("üèÅ FINAL VERDICT:", level_icons[report_obj.risk_level.value])
        print("="*70)

        # Save to export list
        reports.append({
            "sample_index": i,
            "vin": report_obj.vin,
            "risk_level": report_obj.risk_level.value,
            "score": report_obj.score,
            "confidence": report_obj.confidence,
            "suggested_action": report_obj.suggested_action,
            "reasons": report_obj.reasons,
            "details": report_obj.details,
            "timestamp": report_obj.timestamp
        })

    # Export combined JSON report
    out_fname = "auction_report.json"   # <-- fixed filename
    with open(out_fname, "w", encoding="utf-8") as f:
        json.dump(reports, f, indent=2)

    print("="*70)
    print(f"üìÇ Batch complete ‚Äî exported {len(reports)} reports to {out_fname}")


# Execute the batch
if __name__ == "__main__":
    run_batch(TEMPER_SAMPLES)
```
